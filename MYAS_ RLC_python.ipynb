Lo primero en Python es importar los módulos que utilizaremos.

```
import numpy as np #Módulo para hacer operaciones matematicas con matrices.
from scipy import signal # Módulo LTI. 
import matplotlib # graficar
```

Luego definiremos los valores de las constantes y de las matrices de la representación en espacio de estados.

```
R=10.
L=0.1
C=0.000001
A=np.array([[0, -1/L],[1/C,-1/(R*C)]])
B=np.array([[1/L],[0]])
C=np.array([[0,1]])
D=np.array([[0]])
```

Se crea un sistema LTI (lineal invariante en el tiempo, representación en espacio de estados)

```
sys=signal.lti(A,B,C,D)
```

Simulamos la respuesta ante entrada escalón. Ojo que definimos un parámetro opcional T, que es un vector de tiempo. Le indicamos a la función en que puntos queremos la simulación.

```
t,y=signal.step(sys)#,T=np.linspace(0,0.1,1000))
```

Gráfica de salida.

```
matplotlib.pyplot.plot(t,y)
```
