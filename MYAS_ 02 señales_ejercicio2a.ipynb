```python
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib
```

Ahora viene la parte de las funciones y gráficas

```python
# variable tiempo
t = np.arange(-0.01, 5.0, 0.01) # En matlab sería algo así: tt=-1:0.01:10,
#H = np.heaviside(t,1.0) # Esto es la función escalón unitario

u= np.heaviside(t,1.0)+(t-2)*np.heaviside(t-2,1.0)-(t-3)*np.heaviside(t-3,1.0) # Acá se define la señal

# Gráfico
fig, ax = plt.subplots(figsize=(8,4)) # se crea la figura en blanco. Esto permite crear más de una gráfica en el mismo lienzo y modificar las propiedades de cada una
ax.plot(t, u,label='u(t)')

legend = ax.legend()

ax.set(xlabel='tiempo (s)', ylabel='Amplitud',
       title='u(t)')

ax.grid()
plt.ylim([0,5])
plt.show()
```
