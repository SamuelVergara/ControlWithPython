```python
from scipy import signal
import numpy as np
from matplotlib import pyplot
```

```python
Tm=1e-3
```

```python
H1c=signal.lti([1],[1,10])
t,y=H1c.step()
```

```python
pyplot.plot(t,y)
pyplot.grid()
```

```python
K=(1-np.exp(-10*Tm))/10
H1d=signal.dlti([K],[1,-np.exp(-10*Tm)],dt=Tm)
```

```python
t2,y2=signal.dstep(H1d,n=700)
```

```python
pyplot.step(t2, np.squeeze(y2))
pyplot.plot(t,y)
pyplot.grid()
```

```python
H2c=signal.lti([-2],[-4,-5],1)
```

```python
t3,y3=signal.step(H2c)
pyplot.plot(t3,y3)
pyplot.grid()
```

```python
H2d=signal.dlti([np.exp(-2*Tm)],[np.exp(-4*Tm),np.exp(-5*Tm)],9.965e-4)
t4,y4=signal.dstep(H2d,n=1750)
pyplot.step(t4,np.squeeze(y4))
pyplot.grid()
```
