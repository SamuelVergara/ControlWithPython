```python
import numpy as np
import matplotlib.pyplot as plt
```

Ahora viene la parte de las funciones y gráficas

```python
# Data for plotting
t = np.arange(-1.0, 3.0, 0.01)

f = np.heaviside(t,1.0)

alpha=-3.0
beta=2.0
g= f*alpha+beta

fig, ax = plt.subplots()
ax.plot(t, f,label='f(t)')
ax.plot(t, g,label='g(t)')
legend = ax.legend()
#legend = ax.legend(loc='upper left')
ax.set(xlabel='tiempo (s)', ylabel='Amplitud',
       title='f(t) y g(t)')
ax.grid()

plt.show()
```
