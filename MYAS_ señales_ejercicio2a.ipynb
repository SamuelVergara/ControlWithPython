```python
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib
```

Ahora viene la parte de las funciones y gráficas

```python
# Data for plotting
t = np.arange(-0.01, 5.0, 0.01) # tt=-1:0.01:10,
#H = np.heaviside(t,1.0)

u= np.heaviside(t,1.0)+(t-2)*np.heaviside(t-2,1.0)-(t-3)*np.heaviside(t-3,1.0)

fig, ax = plt.subplots()
ax.plot(t, u,label='u(t)')

legend = ax.legend()

ax.set(xlabel='tiempo (s)', ylabel='Amplitud',
       title='u(t)')

ax.grid()
plt.ylim([0,5])
plt.show()
```
