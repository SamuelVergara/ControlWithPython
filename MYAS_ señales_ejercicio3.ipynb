```python
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib
```

Ahora viene la parte de las funciones y gráficas

```python
# Data for plotting
#t = np.arange(-4, 4, 0.01)
#t=t.round(decimals=2)
t=np.linspace(-4,4,num=801)
#H = np.heaviside(t,1.0)

u= 2*np.heaviside(t,1.0)-2*np.heaviside(t-2,1.0)
umenos=2*np.heaviside(-t,1.0)-2*np.heaviside(-t-2,1.0)
up=(u+umenos)/2
ui=(u-umenos)/2
fig, ax = plt.subplots()
#ax.plot(t, u,label='u(t)')
ax.plot(t, up,label='up(t)')
ax.plot(t, ui,label='ui(t)')
legend = ax.legend()
#legend = ax.legend(loc='upper left')
ax.set(xlabel='tiempo (s)', ylabel='Amplitud',
       title='u(t)')

ax.grid()

plt.show()
```
