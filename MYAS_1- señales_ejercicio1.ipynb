## Contextualización de la temática

En modelación y análisis de sistemas dinámicos es habitual utilizar **señales de prueba** para excitar un sistema y estudiar su comportamiento. En este contexto, comprender cómo una señal se modifica bajo **transformaciones matemáticas** es fundamental para interpretar correctamente retardos, cambios de escala, inversiones, ganancias y sesgos.

En esta unidad se trabajará **exclusivamente en tiempo continuo**, utilizando la notación \( u(t) \) para representar una señal genérica. Las transformaciones consideradas se dividen en dos grandes grupos:

- **Transformaciones sobre la variable dependiente** (amplitud).  
- **Transformaciones sobre la variable independiente** (tiempo).

Estas operaciones permiten simular efectos reales como amplificación, desplazamiento de nivel, retardos, compresión temporal y simetrías, ampliamente utilizados en análisis de sistemas, identificación y control.

### A.1 Escalón unitario \(u(t)\)

**Definición matemática**

$$
\mathcal{H}(t) =
\begin{cases}
0, & t < 0 \\
1, & t \ge 0
\end{cases}
$$

````python
## Ejemplo de gráfica de la función escalón unitario usando Matplotlib y NumPy

import numpy as np # Importa la biblioteca NumPy para cálculos numéricos
import matplotlib.pyplot as plt # Importa la biblioteca Matplotlib para gráficos

# Grilla temporal (aproximación a tiempo continuo)
t0 = -10.0 # Tiempo inicial
tf = 10.0 # Tiempo final
N = 5000 # Número de muestras
t = np.linspace(t0, tf, N) # genera un vector desde t0 a tf con N muestras a igual distancia
u=np.heaviside(t,1) # Función escalón unitario

fig, ax = plt.subplots(figsize=(8,4)) # Crea una figura y un eje
ax.plot(t,u,label=r'$\mathcal{H}(t)$') # Grafica la función escalón unitario
ax.set_title('Función escalón unitario') # Título de la gráfica
ax.set_xlabel('Tiempo [s]') # Etiqueta del eje x
ax.set_ylabel('Amplitud') # Etiqueta del eje y
ax.legend() # Leyenda de la gráfica
ax.grid() # Activa la grilla
plt.show() # Muestra la gráfica
````

### Rampa

r(t)=

t, & t \ge 0

### Señal senoidal

**Definición matemática**

$$
u(t) = A \sin\!\left(2\pi f t + \varphi\right) = A \sin\!\left(\omega t + \varphi\right) 
$$

Donde $\omega$ es la frecuencia angular (rad/s) y $f$ la frecuencia en (Hz).

````python
A=1 # Amplitud unitaria
f=1 # Frecuencia de 1 Hz
phi=0 # Fase inicial cero

u_sine=A * np.sin(2*np.pi*f*t + phi)
fig, ax = plt.subplots(figsize=(8,4)) # Crea una figura y un eje
ax.plot(t,u_sine,label=r'$u_{sin}(t)$') # Graf
ax.set_title('Función senoidal') # Título de la gráfica
ax.set_xlabel('Tiempo [s]') # Etiqueta del eje x
ax.set_ylabel('Amplitud') # Etiqueta del eje y
ax.legend() # Leyenda de la gráfica
ax.grid() # Activa la grilla        
plt.show() # Muestra la gráfica
````

````python
# Data for plotting
t = np.arange(-1.0, 3.0, 0.01)

f = np.heaviside(t,1.0)

alpha=-3.0
beta=2.0
g= f*alpha+beta

fig, ax = plt.subplots()
ax.plot(t, f,label='f(t)')
ax.plot(t, g,label='g(t)')
legend = ax.legend()
#legend = ax.legend(loc='upper left')
ax.set(xlabel='tiempo (s)', ylabel='Amplitud',
       title='f(t) y g(t)')
ax.grid()

plt.show()
````
