```python
import numpy as np
from scipy import signal as sg
from matplotlib import pyplot as plt
```

```python
sistemas=[]
for tau in np.linspace(0.1,2,10):
  #se define sistema a partir de coeficientes de polinomios de numerador y denominador
  sistemas.append(sg.lti([1],[tau,1])) 
tt=np.linspace(0,5,1000)
```

```python
for sistema in sistemas:
  t,y =sg.step(sistema,T=tt)
  plt.plot(t,y)
plt.grid()
plt.show()
```
