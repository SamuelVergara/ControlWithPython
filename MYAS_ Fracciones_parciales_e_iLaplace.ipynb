```python
import sympy as sym
```

```python
# INGRESO
#s= sym.Symbol('s')
s, t = sym.symbols('s,t')

Ys=3*(s+1)*(s+3)/(s*(s+2)*(s+4))
#Ys =10*(s+2)/(s*(s**2+8*s+15))
#Ys =7*(s**2+1)/(s**2*(s+3)*(s+4))
#Ys=1/(s+1)**2/(s+2)**2
#Ys=1/s/(s**2+1)
```

Fracciones parciales

```python
pf=sym.apart(Ys)
print(pf)
```

Inversa de Laplace

```python
u=s*0
for exp in pf.args:
  u=u+sym.inverse_laplace_transform(sym.apart(exp),s,t)
print(u)
```
