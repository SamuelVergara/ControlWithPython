```python
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib
```

Ahora viene la parte de las funciones y gráficas

```python
# Data for plotting
t = np.arange(-1.0, 8.0, 0.01) # tt=-1:0.01:10,
#H = np.heaviside(t,1.0)

u= 3*np.heaviside(t-1,1.0)-6*(t-2)*np.heaviside(t-2,1.0)+4*(t-4)*np.heaviside(t-4,1.0)+2*(t-6)*np.heaviside(t-6,1.0)

fig, ax = plt.subplots()
ax.plot(t, u,label='u(t)')

legend = ax.legend()

ax.set(xlabel='tiempo (s)', ylabel='Amplitud',
       title='u(t)')

ax.grid()

plt.show()
```
